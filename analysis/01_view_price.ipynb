```python
import ccxt
import pandas as pd
import matplotlib.pyplot as plt

# 업비트 연결 (읽기 전용)
exchange = ccxt.upbit()

# BTC/KRW 1분봉 최근 200개
ohlcv = exchange.fetch_ohlcv(
    symbol="BTC/KRW",
    timeframe="1m",
    limit=200
)

# DataFrame으로 변환
df = pd.DataFrame(
    ohlcv,
    columns=["ts", "open", "high", "low", "close", "volume"]
)

# 시간 컬럼 변환
df["datetime"] = pd.to_datetime(df["ts"], unit="ms")

df.head()
```
